```python
import networkx as nx 
import numpy as np 
```

```python
def generate_children(d, j, blocks):
    children = []
    end = []
    for x in range(j):
        end.append(0)
    if blocks == end:
        end[0] = 1
        return [end]
    for i in range(j ):
        if blocks[i] + 1 > blocks[i-1] and i != 0:
            continue 
        if blocks[i] < d and blocks[i] != 0:
            new_blocks = blocks.copy()
            new_blocks[i] += 1
            children.append(new_blocks)
        elif blocks[i] == 0 and blocks[i - 1] != 0:
            new_blocks = blocks.copy()
            new_blocks[i] += 1
            children.append(new_blocks)    
    return children
```

```python
def add_edges_recursive(graph, blocks, d, j):
    end = []
    for i in range(j):
        end.append(d)
    if blocks == end:
        return 
    for child in generate_children(d, j, blocks):
        graph.add_edge(tuple(blocks), tuple(child))
        add_edges_recursive(graph, child, d, j)
```

```python
g = nx.DiGraph()
blocks = [0,0,0,0]
d = 4
j = 4
add_edges_recursive(g, blocks, d, j )
# nx.drawing.nx_pydot.write_dot(g, "t.txt")
```

```python
h = nx.DiGraph()
blocks = [0,0,0,0]
d = 2
j = 4
add_edges_recursive(h, blocks, d, j )
nx.drawing.nx_pydot.write_dot(h, "f.txt")
```

```python
def print_pattern(pattern):
    max_height = max(pattern)
    
    for i in range(1, max_height + 1):
        row = ''.join('[] ' if num >= i else '    ' for num in pattern)
        print(row.rstrip())
    
```

```python
for n in g:
    print("******")
    print_pattern(n)
```
