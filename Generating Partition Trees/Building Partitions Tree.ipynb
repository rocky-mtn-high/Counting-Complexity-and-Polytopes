```python
import networkx as nx 
import numpy as np 
import pydot
import graphviz as gv 
```

```python
# j is size of largest part, d is max number of parts. Or think width, height of rectangle respectively
# this function generates all the children of a particular partition
def generate_children(d, j, blocks):
    #houses partitions obtained by adding a box
    children = []  

    #only necessary for empty partition to unique partition of 1
    end = []
    
    #creates a list of length J of 0's in 'end'?
    for x in range(j):  
        end.append(0)
    
    #handle case of empty partition, just add a box and that's it
    if blocks == end:
        end[0] = 1
        return [end]
    
    #since partitions are stored by their conjugates/number of boxes in each column, we have j entries for each
    for i in range(j ):
        # can't add a box in column i if there are the same number in column i-1 
        if blocks[i] + 1 > blocks[i-1] and i != 0:
            continue 
        #add a box in column i if it has space at the bottom and stays a partition
        if blocks[i] < d and blocks[i] != 0:
            new_blocks = blocks.copy()
            new_blocks[i] += 1
            #add new partition to list 'children'
            children.append(new_blocks)
        #extra case -- we can add a box if there isn't one in that column as long as there's something to the left
        elif blocks[i] == 0 and blocks[i - 1] != 0:
            new_blocks = blocks.copy()
            new_blocks[i] += 1
            children.append(new_blocks)    
    #outputs all partititions generated by box-adding
    return children
```

```python
#create edges for the poset diagram based on children function
def add_edges_recursive(graph, blocks, d, j):
    end = []
    for i in range(j):
        end.append(d)
    # again if we are starting from empty partition -- MISSING EDGE?
    if blocks == end:
        return 
    #otherwise, create edges from each parent to each child
    for child in generate_children(d, j, blocks):
        graph.add_edge(tuple(blocks), tuple(child)) # do nodes need to be in tuple form?
        add_edges_recursive(graph, child, d, j)
    
```

```python
k = nx.DiGraph()
blocks = [0,0]
d = 2
j = 2
add_edges_recursive(k, blocks, d, j )
nx.drawing.nx_pydot.write_dot(k, "2box.txt")
```

```python
g = nx.DiGraph()
blocks = [0,0,0]
d = 3
j = 3
add_edges_recursive(g, blocks, d, j )
#nx.drawing.nx_pydot.write_dot(g, "t3.txt")
nx.drawing.nx_pydot.write_dot(g, "3box.txt")
```

```python
def print_pattern(pattern):
    max_height = max(pattern)
    
    for i in range(1, max_height + 1):
        row = ''.join('[] ' if num >= i else '    ' for num in pattern)
        print(row.rstrip())
    
```
